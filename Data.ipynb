```python
df= pd.read_excel('retail-supply-chain-sales-dataset.xlsx',sheet_name ='Retails Order Full Dataset')
df.to_csv('retail-supply-chain-sales-dataset.csv', index= False)
df.head(5)
```

```python
df.drop(columns='Row ID', inplace=True)
df.head(3)
```

```python
df.isnull().sum()
```

```python
df['Order Date']= pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df.dtypes
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
#Handle the missing values
df.fillna({'City': 'Unknown', 'Discount': 0}, inplace=True)
df.dropna(thresh=5, inplace=True)
```

```python
#Handling Duplicates
df.duplicated().sum()
df.drop_duplicates(inplace = True)
```

```python
#Column Names
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
df['profit'] = pd.to_numeric(df['profit'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')
```

```python
df.head(3)
```

```python
df['sale_price'] = df['sales'] - df['discount_amount']
df['profit_margin_%'] = (df['profit'] / df['sales']) * 100
df['delivery_days'] = (df['ship_date'] - df['order_date']).dt.days
df['discount_amount'] = df['sales'] * (df['discount'] / 100)
```
